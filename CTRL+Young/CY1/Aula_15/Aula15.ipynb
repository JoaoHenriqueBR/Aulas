# Aula 15 - Programando Exceções

- **TRY** - Tenta fazer algo
- **EXCEPT** - Excessão (Se algo deu ruim)
- **ELSE** - Se não deu erro
- **FINALLY** - Roda tudo independente do erro

```python
def divide():
    contador = 1
    while True:
        try:
            x = input("Digite um número: ")
            y = input("Digite outro número: ")
            resultado = int(x) / int(y)
        except ZeroDivisionError:
            print("Não é possivel dividir por zero")
            continue
        except ValueError:
            print("Valor inválido! Tente um número inteiro.")
            continue
        else:
            print(resultado)
            break
        finally:
            print("Tentativa número: ", contador)
            contador += 1
```

```python
divide()
```
