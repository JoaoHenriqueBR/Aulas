## Classe

> *É um formato a ser seguido sempre que eu precisar fazer objetos do mesmo tipo, ou que tenham as mesmas caracteristicas ou que tenham o mesmo comportamento.*

**Ex:** Forma de biscoito, planta baixa de uma casa.

- *Atributos:* Características que tenho (`Imobiliaria, Rua, CEP, bairro, número`)
- *Métodos:* Coisas que posso fazer (`enderecoCompleto(), getRua(), setRua(x)`)

```python
class Casa():
    # Atributo da classe
    imobiliaria = "JH Constructions"

    # Método inicializador
    def __init__(self, rua, bairro, CEP, numero):
        self.rua = rua
        self.bairro = bairro
        self.CEP = CEP
        self.numero = numero

    # Métodos getters
    def getImobiliaria(self):
        return self.imobiliaria

    def getRua(self):
        return self.rua

    def getBairro(self):
        return self.bairro

    def getCEP(self):
        return self.CEP

    def getNumero(self):
        return self.numero

    def setImobiliaria(self, i):
        self.imobiliaria = i

    # Métodos setters
    def setRua(self, r):
        self.rua = r

    def setBairro(self, b):
        self.bairro = b

    def setCEP(self, c):
        self.CEP = c

    def setNumero(self, n):
        self.numero = n

    # Método para retornar o endereço completo com base nos atributos
    def enderecoCompleto(self):
        return f"Endereço Completo: {self.rua}, {self.numero}. {self.bairro} - CEP: {self.CEP}"
```

## Objeto

*"Coisa **material** ou **abstrata** que é feita por um **modelo (classe)** e pode ser descrita por meio das suas características **(atributos),** comportamentos **(métodos)** e **estado** atual."*

### Instanciamento

- Quando ocorre uma cópia de um objeto, formado segundo o molde definido pela classe.

> Objeto é uma instância de uma classe


```python
# Primeiro objeto (casa1)
casa1 = Casa(rua="Rua abc da silva", numero=123, CEP=40028922, bairro="Interlagos")

# Segundo objeto (casa2)
casa2 = Casa(rua="Rua qwerty", numero=158, CEP=896749, bairro="Mogi das Cruzes")
```

```python
print(casa1.enderecoCompleto())
print(casa2.enderecoCompleto())
```

#### Atributos de classe

- Atributos que valem para todos os objetos de uma classe.

```python
print(casa1.imobiliaria)
print(casa2.imobiliaria)
```

## Métodos getters e setters

Métodos responsáveis por atualizar e retornar os valores de cada atributo.

- `get` -> Método que retorna
- `set` -> Método que atualiza

```python
casa1.setImobiliaria("PROFESSOR LINDO")
casa1.getImobiliaria()
```
