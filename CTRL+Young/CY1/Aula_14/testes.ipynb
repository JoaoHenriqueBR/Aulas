```python
from modulo import sobrenomeNaOrdem
import unittest

class NomeTest(unittest.TestCase):
    def test_sobrenomeNaOrdem(self):
        nomeCompleto = sobrenomeNaOrdem("João", "Madureira", "Silva")
        self.assertEqual(nomeCompleto, "João Silva Madureira")

    def test_sobrenomeNaOrdem2(self):
        nomeCompleto = sobrenomeNaOrdem("João", "Dias", "Saraiva")
        self.assertEqual(nomeCompleto, "João Saraiva Dias")
```

```python
unittest.main(argv=[''], exit=False)
```
