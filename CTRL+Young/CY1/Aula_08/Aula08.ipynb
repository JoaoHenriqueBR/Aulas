## Abrindo um arquivo

Uma vez tendo o arquivo, só abrir com a função `open()`.

Onde:

```
variavel = open('nome_do_arquivo.txt', 'parametro')
```

O arquivo pode ser aberto com diferentes parâmetros, sem o parâmetro, ele adota o r (Apenas Leitura) como padrão, podemos usar outros, como mostrado abaixo:


- **r (read only)** - Modo de leitura usado quando o arquivo vai apenas ser lido;
    - Se você não especificar qual o modo que vai usar, Python irá assumir que o arquivo será utilizado apenas para leitura;
- **w (write only)** - Modo de gravação que é usado para editar e gravar novas informações no arquivo;
Quaisquer arquivos existente com o mesmo nome serão apagados quando este modo for ativado;
- **a (append)** - Modo de acréscimo, que é usado para adicionar novos dados ao final do arquivo, sendo inserida automaticamente;
- **r+ ou w+ (read, write and append)** - Modo especial de leitura e gravação, usado para lidar com as duas operações ao trabalhar um arquivo;

````python
arquivo = open('teste1.txt', 'r+')
````

## Lendo e Escrevendo

### Lendo

Com a função `read()`, você pode ver todas as linhas de um arquivo.

````python
arquivo.read()
````

Mas o que acontece se você tentar ler o arquivo novamente?

````python
arquivo.read()
````

Isso acontece, como você pode imaginar, porque o *"cursor"* de leitura está no **final** do arquivo após o ter lido. Portanto, não há nada a ler. Você pode redefinir o *"cursor"* para o início do arquivo com o seguinte comando:

````python
# Procure o inicio do arquivo (indice 0)
arquivo.seek(0)

# Agora você pode ler novamente o arquivo
arquivo.read()
````

Você reparou no `\n`? Por padrão, o arquivo é mostrado em uma linha só, onde o `n` representa uma nova linha.

- Mas, e se eu quiser ler o arquivo inteiro linha por linha?
- Podemos fazer isso de duas formas:
    - `readline()` - Função que lê uma linha cada vez.
    - `for linha in arquivo` - Mostra cada linha até chegar ao final do arquivo.

````python
# Método Manual
# com readline()

# Lendo a primeira linha
print(arquivo.readline())
# Lendo a segunda linha
print(arquivo.readline())
# Lendo a terceira linha
print(arquivo.readline())
````

````python
# Método automático
# Com o for

# Lendo o arquivo inteiro com o for
for linha in arquivo:
    print(linha)
````

**Fechando o arquivo**

````python
arquivo.close()
````

### Escrevendo

Como visto antes, usando a função `open()` sem o parâmetro do modo, você só conseguirá ler o arquivo.

Será preciso passar o argumento **“w”**, **"w+"** ou **“r+”** para escrever sobre o arquivo.

- O argumento **"w"** permite **apenas escrever** um **novo conteúdo** do zero ao arquivo até ele ser fechado.
- O argumento **"w+"** permite **escrever e adicionar mais conteúdo** ao arquivo sem ele ser fechado.

````python
arquivo = open('teste1.txt', 'r+')

lista = ['Pizza', 'Baião de Dois', 'Virado a Paulista', 'Cuscuz Paulista']
````

Podemos usar de outros comandos que vimos antes para interagir com os arquivos

````python
for i in lista:
    arquivo.write(i)
````

Usar o **"r+"** permite a leitura do arquivo, além de escrever no arquivo, como o **"w+"**.

````python
arquivo.seek(0)
arquivo.read()
````

````python
arquivo.close()
````

## Comando With

- Fornece uma sintaxe mais limpa
- Todos os arquivos abertos são fechados automaticamente

Sintaxe:

```
with {comando} as {variavel}:
    função(parâmetros)
```

- Exemplos prático com arquivos

````python
# Cria um arquivo, abre em modo de escrita, escreve texto e fecha de uma só vez
with open("exemplo 2.txt", "w") as arquivo:
  arquivo.write("Olá, Mundo!\n")
  arquivo.write("Escrevendo mais uma linha")
````

````python
# Uso de listas para adicionar várias linhas de uma só vez
linhas = ["Linha 1\n", "Linha 2\n", "Linha 3\n"]
with open("exemplo_linhas.txt", "w") as arquivo:
  arquivo.writelines(linhas)
````

````python
# Leitura de Arquivo
with open("exemplo_linhas.txt", "r") as arquivo:
  arquivo.seek(0)
  print(arquivo.read())
````
