##  Aula 7 - Funções

- Muitos comandos do Python são funções, como o print();
- Podemos criar as nossas próprias funções;
- Elas permitem reutilizar qualquer parte do código a qualquer momento.

>Ex: Criar uma função chamada ronaldo(), que toda vez que eu chamo ela pelo comando ronaldo(), escreve para mim o nome do ronaldo sem precisar rescrever o mesmo código.

```python
def ronaldo(msg=''):
    """
    :return: Mostra o nome do Ronaldo
    """
    return 'Ronaldo'+ msg
```

```python
ronaldo(" Bom")
```

```python
ronaldo()
```

- Uma função pode receber **parâmetros** (valores de fora), para realizar diferentes comandos

```python
def menor(a,b): # Função para verificar qual o menor valor entre dois.
    if a <= b:
        return a
    else:
        return b
```

```python
a = 5
b = 1

print('o menor valor entre',a,'e',b,'é',menor(a,b))
```

- Podemos acrescentar valores indefinidos se marcarmos um parâmetro com *.
- Cada valor será acrescentado como **item** em uma **tupla** chamada **ingredientes**

```python
def acai(*ingredientes, tamanho='medio'):
    print('\n Preparando um Açai', tamanho, "com os seguintes ingredientes para o ", ronaldo())

    for i in ingredientes:
        print('-',i)
```

```python
acai("Leite em Pó", "Banana", "Maçã")
```

- Podemos adicionar funções dentro de funções e até mesmo chamar nossa própria função
- Funções usadas dentro de outra formam o que chamamos de **funções compostas**
- Usar a própria função dentro dela mesmo forma uma **função recursiva**

```python
def dobraLencol(lencol, gaveta):
    if lencol < gaveta:
        return 0
    else:
        return 1 + dobraLencol(lencol/2, gaveta)

dobraLencol(200, 25)
```
