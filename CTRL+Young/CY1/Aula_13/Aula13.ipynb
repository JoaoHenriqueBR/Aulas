# Pilares do POO

## Abstração

- Representação de um objeto real.
- Constituido de 3 caracteristicas:
    - Identidade;
    - Características (Atributos);
    - Ações (Métodos).


## Encapsulamento

- Esconde um atributo, tornando o privado a classe que foi criado.
    -  Para tornar um atributo privado, use: `__atributo = valor`
- Melhor segurança e organização

```python
class A:
    a = 1 # atributo público
    __b = 2 # atributo privado a class A

class B(A):
    __c = 3 # atributo privado a B

    def __init__(self):
        print(self.a)
        print(self.__c)

a = A()
print(a.a)
```

**Objeto B** não consegue acessar o atributo `b`, pois é privado apenas ao **Objeto A.**

```python
b = B()
print(b.__b)
print(b.__c)
```

## Herança

Usar características ou atributos de uma classe mãe, ou mais, em outras classes filhas para assim poupar código e agilizar trabalho. Ex:

- `Cachorro` herda `come()` da classe-mãe `AnimalSelvagem()`.
- `Cachorro` herda `getDono()` da classe-mãe `AnimalDomestico()`.

```python
class AnimalSelvagem():
    def mover(self):
        print("estou correndo")

    def come(self):
        print("Comendo")

class AnimalDomestico():
    def mover(self):
        print("estou andando")

    def getDono(self):
        return self.dono

class Cachorro(AnimalSelvagem, AnimalDomestico):
    def __init__(self, dono):
        self.dono = dono

    def late(self):
        print("Auau!")
```

```python
c = Cachorro("João")
```

```python
c.getDono()
```

```python
c.come()
```

```python
c.late()
```
